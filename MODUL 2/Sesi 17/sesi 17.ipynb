# ETL (Extract, Transform, Load) Process
ETL adalah proses yang digunakan untuk mengumpulkan data dari berbagai sumber, mengubahnya ke dalam format yang sesuai, dan memuatnya ke dalam sistem penyimpanan data seperti data warehouse.

Kenapa ETL penting? 
1. **Integrasi Data**: ETL memungkinkan penggabungan data dari berbagai sumber yang berbeda.
2. **Kualitas Data**: Proses transformasi membantu membersihkan dan memperbaiki data.
3. **Analisis Data**: Data yang telah diolah dapat digunakan untuk analisis bisnis yang lebih baik.
  

## Konsep Dasar ETL
1. **Extract (Ekstraksi)**: Mengambil data dari berbagai sumber seperti database, file, API, dll.
2. **Transform (Transformasi)**: Mengubah data menjadi format yang sesuai, termasuk pembersihan data, penggabungan, dan penghitungan.
3. **Load (Pemuatan)**: Memuat data yang telah diubah ke dalam sistem penyimpanan data tujuan.

### Proses Extract (detail)
1. **Identifikasi sumber data**: Menentukan dari mana data akan diambil.
2. **Koneksi ke sumber data**: Membangun koneksi ke sumber data, seperti database atau API.
3. **Pengambilan data**: Mengambil data dari sumber yang telah teridentifikasi dan terhubung.

```python
def get_data(code):

    # Create a connection to the database
    connection = pymysql.connect(
        host=db_host,
        user=db_user,
        password=db_password,
        database=db_name
    )

    cursor = connection.cursor()

    ## Query dari tabel products
    query = '''
    SELECT * FROM products;
    '''
    cursor.execute(query)
    hasil = cursor.fetchall()

    columns_name = [desc[0] for desc in cursor.description]
    products = pd.DataFrame(hasil, columns=columns_name)

    return products
```

### Proses Transform (detail)
1. **Pembersihan data**: Menghapus data duplikat, menangani nilai yang hilang, dan memperbaiki kesalahan data.
2. **Penggabungan data**: Menggabungkan data dari berbagai sumber menjadi satu set data yang konsisten.
3. **Penghitungan dan agregasi**: Melakukan perhitungan atau agregasi data sesuai kebutuhan analisis.


- Merging Data
- Cleaning Data
- Normalizing Data dan Aggregating Data

### Proses Load (detail)
1. **Persiapan target penyimpanan**: Menyiapkan database atau data warehouse untuk menerima data.
2. **Memuat data**: Memasukkan data yang telah diubah ke dalam target penyimpanan.
3. **Verifikasi pemuatan**: Memastikan bahwa data telah dimuat dengan benar dan lengkap.

## ETL Process
### Load data volume besar dengan chunking
Untuk memuat data yang cukup besar, kita dapat menggunakan metode chunking pada pandas. Metode ini memungkinkan kita untuk membaca dan memproses data dalam potongan-potongan kecil, sehingga mengurangi penggunaan memori.
### Ketidak konsistenan data
Ketidak konsistenan data dapat terjadi ketika data berasal dari berbagai sumber yang berbeda. Misalnya, format tanggal yang berbeda, penggunaan satuan yang berbeda, atau penamaan kolom yang tidak konsisten. Untuk mengatasi masalah ini, kita perlu melakukan pembersihan dan transformasi data sebelum memuatnya ke dalam sistem penyimpanan data.
### Perubahan sumber data
Sumber data dapat berubah dari waktu ke waktu, misalnya struktur tabel yang berubah, penambahan atau penghapusan kolom, atau perubahan format data. Untuk mengatasi masalah ini, kita perlu memantau sumber data secara berkala dan memperbarui proses ETL sesuai kebutuhan. 
### Limitasi Tools
Beberapa alat ETL mungkin memiliki keterbatasan dalam hal kapasitas pemrosesan data, dukungan untuk berbagai sumber data, atau kemampuan transformasi data. Untuk mengatasi masalah ini, kita perlu memilih alat ETL yang sesuai dengan kebutuhan bisnis dan memastikan bahwa alat tersebut dapat menangani volume dan kompleksitas data yang dihadapi.

## Optimalisasi proses ETL
Untuk mengoptimalkan proses ETL, kita dapat melakukan beberapa langkah berikut:
1. **Paralelisasi**: Memproses data secara paralel untuk meningkatkan kecepatan pemrosesan.
2. **Penggunaan indeks**: Menggunakan indeks pada database untuk mempercepat query dan pemuatan data.
3. **Pemantauan kinerja**: Memantau kinerja proses ETL secara berkala dan mengidentifikasi bottleneck yang perlu diatasi.
4. **Penggunaan alat yang tepat**: Memilih alat ETL yang sesuai dengan kebutuhan bisnis dan memastikan bahwa alat tersebut dapat menangani volume dan kompleksitas data yang dihadapi.
