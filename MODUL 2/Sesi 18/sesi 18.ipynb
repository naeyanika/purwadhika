# Big Query
BigQuery adalah layanan penyimpanan data berskala besar yang disediakan oleh Google Cloud. Layanan ini memungkinkan pengguna untuk menjalankan kueri SQL pada dataset yang sangat besar dengan kecepatan tinggi. BigQuery dirancang untuk analisis data besar (big data) dan dapat menangani volume data yang sangat besar, mulai dari gigabyte hingga petabyte.

```python
# Library yang dibutuhkan
# ! pip install  pandas-gbq
```

```python
# Load Library
from google.oauth2 import service_account
import pandas_gbq
import pandas as pd
```

```python
# Load Data
data = pd.read_csv('belanjakita.csv')
data
```

```python
# Konfigurasi ke google big query
project_id = 'swift-arcadia-432002-f2'
dataset_id = 'belanjakita'
table_name = 'transactions'
key_path = 'swift-arcadia-432002-f2-993877ab2832.json'

# Set Up credentials
scopes = ['https://www.googleapis.com/auth/bigquery']
credentials = service_account.Credentials.from_service_account_file(key_path, scopes=scopes)
```

```python
# Upload data ke bigquery
pandas_gbq.to_gbq(
    data,
    f'{dataset_id}.{table_name}', 
    project_id=project_id,
    if_exists='replace',
    credentials=credentials
)
```

```python
# Query dari python

query = f'SELECT * FROM `{project_id}.{dataset_id}.{table_name}` LIMIT 10'

df = pandas_gbq.read_gbq(
    query,
    project_id=project_id,
    credentials=credentials
)
df
```
