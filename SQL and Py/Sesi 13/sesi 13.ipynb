# Association (Asosiasi Data)

## üìö Daftar Materi

### 1) Relationship and Types of Variable (Hubungan dan Jenis Variabel)

**Apa itu Variabel?**
- Variabel = data yang nilainya bisa berbeda-beda
- Contoh: tinggi badan, umur, gaji, suhu

**Jenis Variabel:**
- **Numerical (Angka)**: 25 tahun, 170 cm, Rp 5.000.000
- **Categorical (Kategori)**: Laki-laki/Perempuan, Merah/Biru/Hijau

**Hubungan antar Variabel:**
- Bagaimana satu variabel mempengaruhi variabel lain
- Contoh: "Apakah ada hubungan antara tinggi badan dengan berat badan?"

---

### 2) Type of Association based on variable (Jenis Asosiasi berdasarkan Variabel)

**A. Asosiasi Positif** ‚ûï
- Kalau variabel A naik ‚Üí variabel B juga naik
- Contoh: Jam belajar vs Nilai ujian

**B. Asosiasi Negatif** ‚ûñ  
- Kalau variabel A naik ‚Üí variabel B turun
- Contoh: Harga barang vs Jumlah pembeli

**C. Tidak Ada Asosiasi** ‚ö™
- Variabel A dan B tidak saling mempengaruhi
- Contoh: Ukuran sepatu vs Nilai matematika

---

### 3) How to Explore the Relationship? (Cara Mencari Hubungan)

**Langkah-langkah:**

1. **Visualisasi Data** üìä
   - Scatter plot untuk data numerik
   - Bar chart untuk data kategori
   - Heatmap untuk korelasi

2. **Hitung Korelasi** üî¢
   - Angka antara -1 sampai +1
   - +1 = hubungan positif sempurna
   - -1 = hubungan negatif sempurna
   - 0 = tidak ada hubungan

3. **Interpretasi** ü§î
   - Lihat pola dalam grafik
   - Baca nilai korelasi

**Contoh Code:**
```python
import pandas as pd
import matplotlib.pyplot as plt

# Membuat scatter plot
plt.scatter(data['variabel_x'], data['variabel_y'])
plt.xlabel('Variabel X')
plt.ylabel('Variabel Y')
plt.title('Hubungan X vs Y')
plt.show()

# Menghitung korelasi
correlation = data['variabel_x'].corr(data['variabel_y'])
print(f"Korelasi: {correlation}")
```

---

### 4) Type of Numerical Relationship (Jenis Hubungan Numerik)

**A. Linear Relationship** üìà
- Hubungan membentuk garis lurus
- Contoh: Tinggi vs Berat badan

**B. Non-Linear Relationship** üìä
- Hubungan melengkung/tidak lurus
- Contoh: Umur vs Kemampuan belajar

**C. Monotonic Relationship** ‚ÜóÔ∏è
- Selalu naik atau selalu turun (tapi tidak harus lurus)
- Contoh: Pengalaman kerja vs Gaji

**D. No Relationship** ‚ö™
- Tidak ada pola yang jelas
- Data tersebar acak

## üéØ Dua Jenis Hubungan Utama

### 1. **Association ‚Üí Correlation** (Asosiasi ‚Üí Korelasi)
**Definisi:** Hubungan statistik antara dua variabel, tapi **TIDAK** berarti yang satu menyebabkan yang lain.

**Karakteristik:**
- Menunjukkan pola hubungan
- Diukur dengan nilai korelasi (-1 sampai +1)
- **BUKAN** sebab-akibat

**Contoh:**
```python
# Contoh korelasi
import pandas as pd
import numpy as np

# Data tinggi dan berat badan
tinggi = [150, 160, 170, 180, 190]  # cm
berat = [45, 55, 65, 75, 85]        # kg

# Hitung korelasi
correlation = np.corrcoef(tinggi, berat)[0,1]
print(f"Korelasi: {correlation:.2f}")  # Output: 1.00 (korelasi sempurna)
```

### 2. **Causation ‚Üí Regression** (Sebab-Akibat ‚Üí Regresi)
**Definisi:** Hubungan dimana satu variabel **MENYEBABKAN** perubahan pada variabel lain.

**Karakteristik:**
- Ada hubungan sebab-akibat yang jelas
- Dapat digunakan untuk prediksi
- Dianalisis dengan regresi

**Contoh:**
```python
# Contoh regresi
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Data pupuk (X) dan tinggi tanaman (Y)
pupuk = np.array([[1], [2], [3], [4], [5]])      # kg
tinggi_tanaman = np.array([10, 15, 20, 25, 30])  # cm

# Membuat model regresi
model = LinearRegression()
model.fit(pupuk, tinggi_tanaman)

# Prediksi
prediksi = model.predict([[6]])  # Jika pupuk 6 kg
print(f"Prediksi tinggi tanaman: {prediksi[0]:.1f} cm")
```

---

## ‚ö†Ô∏è Perbedaan Penting: Correlation vs Causation

| **Correlation (Korelasi)** | **Causation (Sebab-Akibat)** |
|---------------------------|------------------------------|
| Hubungan statistik | Hubungan sebab-akibat |
| Tidak dapat diprediksi | Dapat diprediksi |
| A dan B bergerak bersama | A menyebabkan B |
| Contoh: Es krim vs Tenggelam | Contoh: Pupuk vs Tinggi tanaman |

### üç¶ Contoh Klasik: Es Krim dan Tenggelam
```
üç¶ Penjualan es krim tinggi ‚ÜîÔ∏è üèä Kasus tenggelam tinggi
```
**Korelasi:** Ada! Keduanya tinggi di musim panas
**Sebab-akibat:** TIDAK! Es krim tidak menyebabkan tenggelam
**Penyebab sebenarnya:** Cuaca panas (variabel ketiga)

---

## üîç Cara Menentukan Jenis Hubungan

### 1. **Untuk Association/Correlation:**
```python
# Scatter plot
plt.scatter(x, y)
plt.xlabel('Variabel X')
plt.ylabel('Variabel Y')
plt.title('Hubungan X vs Y')

# Hitung korelasi

```

### 2. **Untuk Causation/Regression:**
```python
# Regresi linear
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X.reshape(-1,1), y)

# R-squared (seberapa baik model menjelaskan data)

print(f"R-squared: {r2:.2f}")
```

# Response Variable and Explanatory Variable
## (Variabel Respons dan Variabel Penjelas)

---

## üéØ Mengapa Perlu Membedakan Variabel?

Ketika menganalisis hubungan antara dua variabel, kita harus terlebih dahulu membedakan peran masing-masing variabel:
- **Mana yang mempengaruhi?** (Explanatory Variable)
- **Mana yang dipengaruhi?** (Response Variable)

---

## üìä Response Variable (Variabel Respons)

**Definisi:** Variabel yang **dipengaruhi** atau **tergantung** pada variabel lain.

**Karakteristik:**
- Disebut juga **Dependent Variable** (Variabel Terikat)
- Nilainya **bergantung** pada explanatory variable
- Biasanya dilambangkan dengan **Y**
- Yang ingin kita **prediksi** atau **jelaskan**

**Contoh:**
```
üìà Nilai ujian (tergantung pada jam belajar)
‚öñÔ∏è Berat badan (tergantung pada tinggi badan)
üí∞ Gaji (tergantung pada pengalaman kerja)
üå°Ô∏è Suhu ruangan (tergantung pada penggunaan AC)
```

---

## üìã Explanatory Variable (Variabel Penjelas)

**Definisi:** Variabel yang **mempengaruhi** atau **menjelaskan** perubahan pada response variable.

**Karakteristik:**
- Disebut juga **Independent Variable** (Variabel Bebas)
- Nilainya **tidak tergantung** pada variabel lain dalam analisis
- Biasanya dilambangkan dengan **X**
- Yang digunakan untuk **memprediksi** atau **menjelaskan**

**Contoh:**
```
üìö Jam belajar (mempengaruhi nilai ujian)
üìè Tinggi badan (mempengaruhi berat badan)
üéì Pengalaman kerja (mempengaruhi gaji)
‚ùÑÔ∏è Penggunaan AC (mempengaruhi suhu ruangan)
```

---

## üîç Jenis Explanatory Variable

### 1. **Quantitative (Kuantitatif)** üî¢
**Definisi:** Variabel yang dapat diukur dengan angka dan memiliki makna matematis.

**Karakteristik:**
- Nilai berupa angka
- Dapat dihitung rata-rata, median, dll
- Bisa dilakukan operasi matematika

**Contoh:**
```python
# Data kuantitatif
jam_belajar = [2, 4, 6, 8, 10]        # jam per hari
tinggi_badan = [150, 160, 170, 180]   # cm
umur = [20, 25, 30, 35]               # tahun
gaji = [5000000, 7000000, 9000000]    # rupiah
```

### 2. **Qualitative (Kualitatif)** üìù
**Definisi:** Variabel yang berupa kategori atau kelompok.

**Karakteristik:**
- Nilai berupa kategori/label
- Tidak bisa dihitung rata-rata
- Analisis menggunakan perbandingan antar grup

**Contoh:**
```python
# Data kualitatif
jenis_kelamin = ['Laki-laki', 'Perempuan']
pendidikan = ['SD', 'SMP', 'SMA', 'S1', 'S2']
kota_asal = ['Jakarta', 'Bandung', 'Surabaya']
status_kerja = ['PNS', 'Swasta', 'Wiraswasta']
```

**Analisis Kualitatif:**
- Menggunakan fungsi agregat: mean, sum, count, percentage
- Membandingkan rata-rata antar grup
- Menggunakan visualisasi bar chart, pie chart

---

## ‚öñÔ∏è Konsep Causation (Sebab-Akibat)

**Prinsip Penting:**
> "Jika perubahan X menyebabkan perubahan Y, TIDAK berarti perubahan Y menyebabkan perubahan X"

**Penjelasan:**
- Hubungan sebab-akibat **tidak bisa dibalik**
- Explanatory variable adalah **penyebab**
- Response variable adalah **akibat**

**Contoh:**
```
‚úÖ Benar: Pupuk (X) ‚Üí Tinggi tanaman (Y)
‚ùå Salah: Tinggi tanaman (Y) ‚Üí Pupuk (X)

‚úÖ Benar: Jam belajar (X) ‚Üí Nilai ujian (Y)  
‚ùå Salah: Nilai ujian (Y) ‚Üí Jam belajar (X)
```

# Type of Association based on Variable
## (Jenis Asosiasi berdasarkan Tipe Variabel)

---

## üîç Mengapa Tipe Variabel Penting?

Cara kita menganalisis hubungan antara dua variabel **sangat tergantung** pada jenis variabel yang kita miliki. Setiap kombinasi membutuhkan pendekatan dan tools analisis yang berbeda.

---

## üìä Tiga Kemungkinan Kombinasi Variabel

### 1. **Qualitative vs Qualitative** (Kategori vs Kategori)

**Definisi:** Kedua variabel berupa kategori atau kelompok

**Contoh:**
```
üë• Gender dan Education Level (Jenis Kelamin dan Tingkat Pendidikan)
üèôÔ∏è City dan Job Type (Kota dan Jenis Pekerjaan)
üé® Color Preference dan Age Group (Preferensi Warna dan Kelompok Umur)
```

**Metode Analisis:**
- **Contingency Table** (Tabel Kontingensi)
- **Chi-Square Test** (Uji Chi-Square)
- **Cram√©r's V** (mengukur kekuatan asosiasi)

**Contoh Code:**
```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import seaborn as sns

# Data contoh
data = pd.DataFrame({
    'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'] * 50,
    'education': ['High School', 'Bachelor', 'Master', 'Bachelor', 'High School', 'Master'] * 50
})

# Membuat contingency table
contingency_table = pd.crosstab(data['gender'], data['education'])
print("Contingency Table:")
print(contingency_table)

# Chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)
print(f"\nChi-square: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")

# Visualisasi
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues')
plt.title('Hubungan Gender vs Education')
plt.show()
```

---

### 2. **Quantitative vs Qualitative** (Angka vs Kategori)

**Definisi:** Satu variabel berupa angka, satu variabel berupa kategori

**Contoh:**
```
üí∞ Income dan Race (Pendapatan dan Ras)
üìè Height dan Gender (Tinggi Badan dan Jenis Kelamin)
üéì Salary dan Department (Gaji dan Departemen)
```

**Metode Analisis:**
- **Group Comparison** (Perbandingan Grup)
- **ANOVA** (Analysis of Variance)
- **Box Plot** dan **Bar Chart**

**Contoh Code:**
```python
import matplotlib.pyplot as plt
from scipy import stats

# Data contoh
data = pd.DataFrame({
    'height': [165, 170, 160, 175, 168, 172, 162, 178, 166, 174,
               155, 158, 152, 160, 157, 162, 150, 165, 159, 161],
    'gender': ['Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male',
               'Female', 'Female', 'Female', 'Female', 'Female', 'Female', 'Female', 'Female', 'Female', 'Female']
})

# Group statistics
group_stats = data.groupby('gender')['height'].agg(['mean', 'std', 'count'])
print("Group Statistics:")
print(group_stats)

# T-test untuk membandingkan rata-rata
male_height = data[data['gender'] == 'Male']['height']
female_height = data[data['gender'] == 'Female']['height']
t_stat, p_value = stats.ttest_ind(male_height, female_height)
print(f"\nT-test: t={t_stat:.4f}, p={p_value:.4f}")

# Visualisasi
plt.figure(figsize=(10, 6))

# Box plot
plt.subplot(1, 2, 1)
sns.boxplot(x='gender', y='height', data=data)
plt.title('Box Plot: Height by Gender')

# Bar plot (rata-rata)
plt.subplot(1, 2, 2)
sns.barplot(x='gender', y='height', data=data, ci=95)
plt.title('Average Height by Gender')

plt.tight_layout()
plt.show()
```

---

### 3. **Quantitative vs Quantitative** (Angka vs Angka)

**Definisi:** Kedua variabel berupa angka/numerik

**Contoh:**
```
üå°Ô∏è Air Temperature dan Humidity (Suhu Udara dan Kelembaban)
‚öñÔ∏è Weight dan Height (Berat Badan dan Tinggi Badan)
üìö Study Hours dan Exam Score (Jam Belajar dan Nilai Ujian)
```

**Metode Analisis:**
- **Correlation Analysis** (Analisis Korelasi)
- **Scatter Plot**
- **Linear Regression**

**Contoh Code:**
```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Data contoh
np.random.seed(42)
weight = np.random.normal(70, 10, 100)  # kg
height = 100 + 0.8 * weight + np.random.normal(0, 5, 100)  # cm

data = pd.DataFrame({
    'weight': weight,
    'height': height
})

# Correlation
correlation = data['weight'].corr(data['height'])
print(f"Correlation: {correlation:.4f}")

# Linear regression
X = data[['weight']]
y = data['height']
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)

print(f"R-squared: {r2:.4f}")
print(f"Slope: {model.coef_[0]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")

# Visualisasi
```

```python
# Untuk tabel 2x2, hitung odds ratio
def calculate_odds_ratio(table_2x2):
    """
    Menghitung odds ratio dari tabel 2x2
    table_2x2: [[a, b], [c, d]]
    OR = (a*d) / (b*c)
    """
    a, b = table_2x2[0]
    c, d = table_2x2[1]
    
    if b * c == 0:
        return "Cannot calculate (division by zero)"
    
    odds_ratio = (a * d) / (b * c)
    return odds_ratio
```

```python
    subset = data[data[cat_var1] == target_cat1]
```

### 5. **Chi-Square Test** üßÆ
**Definisi:** Uji statistik untuk mengetahui apakah ada hubungan signifikan antara dua variabel kategori

```python
from scipy.stats import chi2_contingency

# Chi-square test
def perform_chi_square_test(data, var1, var2):
    # Buat contingency table
    contingency = pd.crosstab(data[var1], data[var2])
    
    # Lakukan chi-square test
    chi2, p_value, dof, expected = chi2_contingency(contingency)
    
    print("Chi-Square Test Results:")
    print(f"Chi-square statistic: {chi2:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Degrees of freedom: {dof}")
    print(f"Significance level: {'Significant' if p_value < 0.05 else 'Not significant'} (Œ± = 0.05)")
    
    # Cram√©r's V (effect size)
    n = contingency.sum().sum()
    cramers_v = np.sqrt(chi2 / (n * (min(contingency.shape) - 1)))
    print(f"Cram√©r's V (effect size): {cramers_v:.4f}")
    
    return chi2, p_value, dof, expected, cramers_v

# Lakukan test
chi2, p_val, dof, expected, cramers_v = perform_chi_square_test(data, 'gender', 'education')
```

# Correlation (Korelasi)
## (Pengukuran Hubungan Antar Variabel)

---

## üîç Apa itu Correlation?

**Correlation** adalah ukuran statistik yang menunjukkan seberapa kuat hubungan linear antara dua variabel numerik.

---

## üìä Poin-Poin Penting tentang Correlation

### 1. **Correlation is about association** üîó
- Korelasi mengukur **asosiasi** atau hubungan antara variabel
- **Association doesn't imply causation** (Asosiasi ‚â† Sebab-akibat)
- Hanya menunjukkan pola hubungan, bukan penyebab

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Contoh: Es krim vs Kasus tenggelam (korelasi tanpa sebab-akibat)
data_spurious = pd.DataFrame({
    'ice_cream_sales': [100, 150, 200, 250, 300],
    'drowning_cases': [10, 15, 20, 25, 30],
    'temperature': [20, 25, 30, 35, 40]  # Variabel tersembunyi
})

correlation = data_spurious['ice_cream_sales'].corr(data_spurious['drowning_cases'])
print(f"Korelasi Es Krim vs Tenggelam: {correlation:.3f}")
print("Korelasi tinggi, tapi es krim TIDAK menyebabkan tenggelam!")
print("Penyebab sebenarnya: Suhu panas (variabel ketiga)")
```

### 2. **Correlation doesn't differentiate variables** ‚öñÔ∏è
- Korelasi **tidak membedakan** response variable (X) dan explanatory variable (Y)
- **r(X,Y) = r(Y,X)** ‚Üí Hasilnya sama
- Berbeda dengan regresi yang membedakan dependent/independent

```python
# Demonstrasi: korelasi tidak membedakan urutan variabel
data = pd.DataFrame({
    'height': [150, 160, 170, 180, 190],
    'weight': [50, 60, 70, 80, 90]
})

# Korelasi Height ‚Üí Weight
corr1 = data['height'].corr(data['weight'])
# Korelasi Weight ‚Üí Height  
corr2 = data['weight'].corr(data['height'])

print(f"Korelasi Height vs Weight: {corr1:.3f}")
print(f"Korelasi Weight vs Height: {corr2:.3f}")
print(f"Sama? {corr1 == corr2}")  # True
```

### 3. **Correlation only measures strength and direction** üìè
- **Strength (Kekuatan)**: Seberapa kuat hubungan
- **Direction (Arah)**: Positif atau negatif
- **TIDAK mengukur**: Sebab-akibat, prediksi, perubahan

```python
# Contoh berbagai kekuatan korelasi
np.random.seed(42)
n = 100

# Strong positive correlation
x1 = np.random.normal(0, 1, n)
y1 = 2 * x1 + np.random.normal(0, 0.5, n)

# Weak correlation
x2 = np.random.normal(0, 1, n)
y2 = 0.3 * x2 + np.random.normal(0, 2, n)

# Strong negative correlation
x3 = np.random.normal(0, 1, n)
y3 = -2 * x3 + np.random.normal(0, 0.5, n)

correlations = {
    'Strong Positive': np.corrcoef(x1, y1)[0,1],
    'Weak': np.corrcoef(x2, y2)[0,1],
    'Strong Negative': np.corrcoef(x3, y3)[0,1]
}

for desc, corr in correlations.items():
    print(f"{desc}: {corr:.3f}")
```

## üìà Correlation Range dan Interpretasi

### **Range: -1 < r < 1** 
```
r = +1.0  ‚Üí  Perfect Positive Correlation
r = +0.8  ‚Üí  Strong Positive
r = +0.5  ‚Üí  Medium Positive  
r = +0.2  ‚Üí  Weak Positive
r =  0.0  ‚Üí  No Linear Relationship
r = -0.2  ‚Üí  Weak Negative
r = -0.5  ‚Üí  Medium Negative
r = -0.8  ‚Üí  Strong Negative
r = -1.0  ‚Üí  Perfect Negative Correlation
```

### **Klasifikasi Kekuatan:**
- **0 - 0.3**: **Weak** (Lemah)
- **0.3 - 0.7**: **Medium** (Sedang)
- **0.7 - 1**: **Strong** (Kuat)

```python
def interpret_correlation(r):
    """Interpretasi kekuatan dan arah korelasi"""
    abs_r = abs(r)
    
    # Tentukan kekuatan
    if abs_r >= 0.7:
        strength = "Strong"
    elif abs_r >= 0.3:
        strength = "Medium"
    else:
        strength = "Weak"
    
    # Tentukan arah
    if r > 0:
        direction = "Positive"
    elif r < 0:
        direction = "Negative"
    else:
        direction = "No"
    
    return f"{direction} {strength}"

# Test fungsi
test_correlations = [0.85, -0.45, 0.15, -0.92, 0.0]
for corr in test_correlations:
    print(f"r = {corr:5.2f} ‚Üí {interpret_correlation(corr)}")
```

## 1Ô∏è‚É£ QUALITATIVE vs QUALITATIVE
### Contoh: Gender vs Education Level

````python
# Membuat dataset contoh: Gender vs Education
np.random.seed(42)
n = 300

# Buat data dengan sedikit bias (realistis)
gender = np.random.choice(['Male', 'Female'], n, p=[0.6, 0.4])
education = []

for g in gender:
    if g == 'Male':
        # Male lebih banyak di teknik/sains
        edu = np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], 
                              p=[0.3, 0.4, 0.2, 0.1])
    else:
        # Female lebih merata
        edu = np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], 
                              p=[0.25, 0.45, 0.25, 0.05])
    education.append(edu)

data_qual = pd.DataFrame({
    'gender': gender,
    'education': education
})

print("üìä Dataset Gender vs Education:")
print(data_qual.head(10))
print(f"\nUkuran data: {len(data_qual)} baris")
print(f"Gender distribution: \n{data_qual['gender'].value_counts()}")
print(f"\nEducation distribution: \n{data_qual['education'].value_counts()}")
````

````python
# üìã Contingency Table (Tabel Kontingensi)
contingency = pd.crosstab(data_qual['gender'], data_qual['education'], margins=True)
print("üî¢ CONTINGENCY TABLE:")
print("="*50)
print(contingency)

# Dalam bentuk persentase
print("\nüìä CONTINGENCY TABLE (PERCENTAGE):")
print("="*50)
contingency_pct = pd.crosstab(data_qual['gender'], data_qual['education'], normalize='all') * 100
print(contingency_pct.round(2))
````

````python
# üßÆ Chi-Square Test
chi2, p_value, dof, expected = chi2_contingency(contingency.iloc[:-1, :-1])  # Exclude margins

print("üî¨ CHI-SQUARE TEST RESULTS:")
print("="*50)
print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of freedom: {dof}")

# Interpretasi
if p_value < 0.05:
    significance = "üü¢ SIGNIFIKAN (Ada hubungan)"
else:
    significance = "üî¥ TIDAK SIGNIFIKAN (Tidak ada hubungan)"

print(f"Hasil: {significance}")

# Cram√©r's V (effect size)
n = contingency.iloc[:-1, :-1].sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(contingency.iloc[:-1, :-1].shape) - 1)))
print(f"Cram√©r's V (kekuatan hubungan): {cramers_v:.4f}")

# Interpretasi Cram√©r's V
if cramers_v < 0.1:
    strength = "Sangat Lemah"
elif cramers_v < 0.3:
    strength = "Lemah"
elif cramers_v < 0.5:
    strength = "Sedang"
else:
    strength = "Kuat"

print(f"Kekuatan hubungan: {strength}")
````

### Contoh: Salary vs Gender

````python
# Membuat dataset Salary vs Gender
np.random.seed(123)
n = 200

# Generate data dengan sedikit bias gender (realistis)
gender_quant = np.random.choice(['Male', 'Female'], n, p=[0.55, 0.45])
salary = []

for g in gender_quant:
    if g == 'Male':
        # Male rata-rata sedikit lebih tinggi (bias sistemik)
        sal = np.random.normal(8500000, 2000000)  # 8.5 juta ¬± 2 juta
    else:
        sal = np.random.normal(7800000, 1800000)  # 7.8 juta ¬± 1.8 juta
    
    # Pastikan gaji minimum
    sal = max(sal, 4000000)  # Minimum 4 juta
    salary.append(sal)

data_quant_qual = pd.DataFrame({
    'gender': gender_quant,
    'salary': salary
})

print("üí∞ Dataset Salary vs Gender:")
print(data_quant_qual.head(10))
print(f"\nUkuran data: {len(data_quant_qual)} baris")
print(f"\nRingkasan Statistik Gaji:")
print(data_quant_qual.groupby('gender')['salary'].agg(['count', 'mean', 'std', 'min', 'max']).round(0))
````

````python
# üìä T-Test untuk membandingkan rata-rata gaji
male_salary = data_quant_qual[data_quant_qual['gender'] == 'Male']['salary']
female_salary = data_quant_qual[data_quant_qual['gender'] == 'Female']['salary']

# Independent t-test
t_stat, p_value_ttest = ttest_ind(male_salary, female_salary)

print("üß™ INDEPENDENT T-TEST RESULTS:")
print("="*50)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value_ttest:.4f}")

# Interpretasi
if p_value_ttest < 0.05:
    significance_ttest = "üü¢ SIGNIFIKAN (Ada perbedaan rata-rata)"
else:
    significance_ttest = "üî¥ TIDAK SIGNIFIKAN (Tidak ada perbedaan rata-rata)"

print(f"Hasil: {significance_ttest}")

# Effect size (Cohen's d)
pooled_std = np.sqrt(((len(male_salary) - 1) * male_salary.var() + 
                     (len(female_salary) - 1) * female_salary.var()) / 
                    (len(male_salary) + len(female_salary) - 2))

cohens_d = (male_salary.mean() - female_salary.mean()) / pooled_std

print(f"\nCohen's d (effect size): {cohens_d:.4f}")

# Interpretasi effect size
if abs(cohens_d) < 0.2:
    effect = "Kecil"
elif abs(cohens_d) < 0.5:
    effect = "Sedang"
elif abs(cohens_d) < 0.8:
    effect = "Besar"
else:
    effect = "Sangat Besar"

print(f"Ukuran efek: {effect}")

print(f"\nüìà Rata-rata gaji Male: Rp {male_salary.mean():,.0f}")
print(f"üìà Rata-rata gaji Female: Rp {female_salary.mean():,.0f}")
print(f"üí∞ Selisih: Rp {male_salary.mean() - female_salary.mean():,.0f}")
````

## 5Ô∏è‚É£ LATIHAN INTERAKTIF
### Fungsi untuk Interpretasi Hasil

````python
# üîß FUNGSI HELPER untuk Analisis Association

def analyze_qualitative_association(data, var1, var2):
    """
    Fungsi untuk menganalisis hubungan antara dua variabel kategori
    """
    print(f"üîç ANALYZING: {var1} vs {var2}")
    print("="*50)
    
    # Contingency table
    contingency = pd.crosstab(data[var1], data[var2], margins=True)
    print("üìã Contingency Table:")
    print(contingency)
    
    # Chi-square test
    chi2, p_value, dof, expected = chi2_contingency(contingency.iloc[:-1, :-1])
    
    print(f"\nüßÆ Chi-square Test:")
    print(f"Chi-square: {chi2:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Degrees of freedom: {dof}")
    
    # Interpretation
    if p_value < 0.05:
        print("üü¢ Result: SIGNIFICANT association")
    else:
        print("üî¥ Result: NO significant association")
    
    # Cram√©r's V
    n = contingency.iloc[:-1, :-1].sum().sum()
    cramers_v = np.sqrt(chi2 / (n * (min(contingency.iloc[:-1, :-1].shape) - 1)))
    print(f"Cram√©r's V: {cramers_v:.4f}")
    
    return {
        'contingency': contingency,
        'chi2': chi2,
        'p_value': p_value,
        'cramers_v': cramers_v
    }

def analyze_quantitative_association(data, var1, var2):
    """
    Fungsi untuk menganalisis hubungan antara dua variabel numerik
    """
    print(f"üîç ANALYZING: {var1} vs {var2}")
    print("="*50)
    
    # Correlations
    pearson_r, pearson_p = pearsonr(data[var1], data[var2])
    spearman_r, spearman_p = spearmanr(data[var1], data[var2])
    
    print(f"üîó Pearson correlation: {pearson_r:.4f} (p-value: {pearson_p:.4f})")
    print(f"üîó Spearman correlation: {spearman_r:.4f} (p-value: {spearman_p:.4f})")
    
    # Interpretation
    def interpret_corr(r):
        abs_r = abs(r)
        if abs_r >= 0.8:
            strength = "Very Strong"
        elif abs_r >= 0.6:
            strength = "Strong"
        elif abs_r >= 0.4:
            strength = "Moderate"
        elif abs_r >= 0.2:
            strength = "Weak"
        else:
            strength = "Very Weak"
        
        direction = "Positive" if r > 0 else "Negative" if r < 0 else "No"
        return f"{direction} {strength}"
    
    print(f"üìä Interpretation: {interpret_corr(pearson_r)}")
    
    # Linear regression
    from sklearn.linear_model import LinearRegression
    X = data[[var1]]
    y = data[var2]
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    
    print(f"\nüìà Linear Regression:")
    print(f"R-squared: {r2:.4f}")
    print(f"Equation: {var2} = {model.intercept_:.2f} + {model.coef_[0]:.4f} √ó {var1}")
    
    return {
        'pearson_r': pearson_r,
        'pearson_p': pearson_p,
        'spearman_r': spearman_r,
        'spearman_p': spearman_p,
        'r2': r2,
        'slope': model.coef_[0],
        'intercept': model.intercept_
    }

def analyze_mixed_association(data, categorical_var, numerical_var):
    """
    Fungsi untuk menganalisis hubungan antara variabel kategori dan numerik
    """
    print(f"üîç ANALYZING: {categorical_var} vs {numerical_var}")
    print("="*50)
    
    # Group statistics
    group_stats = data.groupby(categorical_var)[numerical_var].agg(['count', 'mean', 'std', 'min', 'max'])
    print("üìä Group Statistics:")
    print(group_stats.round(3))
    
    # Get groups for t-test (assuming 2 groups)
    groups = data[categorical_var].unique()
    if len(groups) == 2:
        group1 = data[data[categorical_var] == groups[0]][numerical_var]
        group2 = data[data[categorical_var] == groups[1]][numerical_var]
        
        # T-test
        t_stat, p_value = ttest_ind(group1, group2)
        
        print(f"\nüß™ Independent T-test:")
        print(f"T-statistic: {t_stat:.4f}")
        print(f"P-value: {p_value:.4f}")
        
        if p_value < 0.05:
            print("üü¢ Result: SIGNIFICANT difference between groups")
        else:
            print("üî¥ Result: NO significant difference between groups")
        
        # Effect size (Cohen's d)
        pooled_std = np.sqrt(((len(group1) - 1) * group1.var() + 
                             (len(group2) - 1) * group2.var()) / 
                            (len(group1) + len(group2) - 2))
        cohens_d = (group1.mean() - group2.mean()) / pooled_std
        print(f"Cohen's d (effect size): {cohens_d:.4f}")
        
        return {
            'group_stats': group_stats,
            't_statistic': t_stat,
            'p_value': p_value,
            'cohens_d': cohens_d
        }
    else:
        print(f"\n‚ö†Ô∏è Note: {len(groups)} groups found. Use ANOVA for multiple groups.")
        return {'group_stats': group_stats}

print("‚úÖ Fungsi helper berhasil dibuat!")
print("üìù Gunakan:")
print("‚Ä¢ analyze_qualitative_association(data, 'var1', 'var2')")
print("‚Ä¢ analyze_quantitative_association(data, 'var1', 'var2')")
print("‚Ä¢ analyze_mixed_association(data, 'categorical_var', 'numerical_var')")
````

## üåä 2. NON-LINEAR or MONOTONIC RELATIONSHIP

### **Karakteristik:**
- Data membentuk **kurva** (bukan garis lurus)
- Hubungan tetap **monotonic** (selalu naik atau selalu turun)
- **Spearman Correlation** lebih efektif daripada Pearson

- **Quadratic** (Kuadratik): Y = aX¬≤ + bX + c
- **Cubic** (Kubik): Y = aX¬≥ + bX¬≤ + cX + d
- **Exponential**: Y = ae^(bX)
- **Logarithmic**: Y = a √ó ln(X) + b

```python
def analyze_relationship_type(x, y, plot=True):
    """
    Fungsi untuk menganalisis jenis hubungan antara dua variabel
    """
    # Calculate correlations
    pearson_r, pearson_p = pearsonr(x, y)
    spearman_r, spearman_p = spearmanr(x, y)
    
    # Determine relationship type
    diff = abs(pearson_r - spearman_r)
    
    if abs(pearson_r) < 0.1 and abs(spearman_r) < 0.1:
        relationship_type = "No Relationship"
        recommendation = "No significant relationship"
    elif diff < 0.1:
        relationship_type = "Linear"
        recommendation = "Use Pearson correlation"
    elif abs(spearman_r) > abs(pearson_r):
        relationship_type = "Non-Linear Monotonic"
        recommendation = "Use Spearman correlation"
    else:
        relationship_type = "Non-Linear Non-Monotonic"
        recommendation = "Use specialized analysis (polynomial regression, etc.)"
    
    print(f"üîç RELATIONSHIP ANALYSIS:")
    print(f"Pearson correlation: {pearson_r:.4f}")
    print(f"Spearman correlation: {spearman_r:.4f}")
    print(f"Difference: {diff:.4f}")
    print(f"Relationship type: {relationship_type}")
    print(f"Recommendation: {recommendation}")
    
    if plot:
        plt.figure(figsize=(10, 6))
        
        plt.subplot(1, 2, 1)
        plt.scatter(x, y, alpha=0.6)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title(f'Scatter Plot\n{relationship_type}')
        plt.grid(True, alpha=0.3)
        
        plt.subplot(1, 2, 2)
        correlations = ['Pearson', 'Spearman']
        values = [pearson_r, spearman_r]
        colors = ['blue', 'red']
        
        plt.bar(correlations, values, color=colors, alpha=0.7)
        plt.ylabel('Correlation Coefficient')
        plt.title('Correlation Comparison')
        plt.ylim(-1, 1)
        
        for i, v in enumerate(values):
            plt.text(i, v + 0.05, f'{v:.3f}', ha='center', va='bottom')
        
        plt.tight_layout()
        plt.show()
    
    return {
        'pearson_r': pearson_r,
        'spearman_r': spearman_r,
        'relationship_type': relationship_type,
        'recommendation': recommendation
    }
```

- H1 : Terdapat hubungan antara x (product category) dan y (payment method)
