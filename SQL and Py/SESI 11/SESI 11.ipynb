# Bar Plot
Bar Plot adalah jenis grafik yang digunakan untuk menampilkan data kategorikal dengan menggunakan batang atau kolom. Setiap batang mewakili kategori tertentu, dan panjang atau tinggi batang tersebut menunjukkan nilai atau frekuensi dari kategori tersebut. Bar Plot sangat berguna untuk membandingkan nilai antar kategori secara visual.
Kapan menggunakan Bar Plot:
1. Data Kategorikal: Bar Plot digunakan ketika data yang ingin ditampilkan adalah kategorikal, seperti jenis produk, wilayah, atau kategori lainnya.
2. Perbandingan Antar Kategori: Bar Plot efektif untuk membandingkan nilai antar kategori, sehingga memudahkan dalam melihat perbedaan atau tren.
3. Data Diskrit: Bar Plot cocok untuk data diskrit, di mana nilai-nilai tertentu memiliki arti khusus dan tidak kontinu.
4. Visualisasi yang Sederhana: Bar Plot memberikan visualisasi yang sederhana dan mudah dipahami, sehingga cocok untuk presentasi data kepada audiens yang tidak terlalu teknis.
Contoh penggunaan Bar Plot:
- Menampilkan jumlah penjualan per kategori produk.
- Membandingkan pendapatan antar wilayah.
- Menunjukkan frekuensi kejadian dalam berbagai kategori.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales = pd.read_csv('sales_data.csv')
```

```python
# Tabel Bantu

reg = sales['region'].value_counts().reset_index()
reg
```

## Barchart

```python
# Barchart

plt.figure(figsize=(5,5))
colors = plt.cm.coolwarm(reg['count'] / reg['count'].max())

plt.bar(x = reg['region'], height = reg['count'], color = colors)
plt.title('Jumlah Order per Region')
plt.xlabel('Region')
plt.ylabel('Jumlah Order')
plt.show()
```

```python
# Barchart (Bar Horizontal)

plt.figure(figsize=(5,5))
colors = plt.cm.coolwarm(reg['count'] / reg['count'].max())

plt.barh(y = reg['region'], width = reg['count'], color = colors)
plt.title('Jumlah Order per Region')
plt.xlabel('Jumlah Order')
plt.ylabel('Region')
plt.show()
```

```python
# Barchart (Seaborn) dengan gradient warna

plt.figure(figsize=(5,5))

sns.barplot(x= 'region', y='price', data=sales, estimator='median', ci=None)
plt.show()
```

```python
# Barchart (Seaborn) dengan gradient warna

plt.figure(figsize=(7,7))
sns.barplot(x='region', y='price', data=sales, estimator='sum', errorbar=('ci', 0))
plt.title('Total Penjualan Penjualan Per Region')
plt.xlabel('Region')
plt.ylabel('Total Penjualan')
plt.grid(axis='y', linestyle='--')

plt.show()
```

```python
# Bar Chart - Seaborn

plt.figure(figsize=(7,7))
bar_plot = sns.barplot(x = 'region',
            y = 'price',
            data = sales,
            estimator = 'sum',
            errorbar=None,
            palette='coolwarm')

for i in bar_plot.containers:
    bar_plot.bar_label(i, fmt='%.0f', padding=3, fontsize=10, fontweight='bold')

plt.title('Total Penjualan Penjualan Per Region')
plt.xlabel('Region')
plt.ylabel('Total Penjualan')
plt.grid(axis = 'y', linestyle = '--')

plt.show()
```

# Piechart
Pie Chart adalah jenis grafik yang digunakan untuk menampilkan proporsi atau persentase dari berbagai kategori dalam bentuk irisan lingkaran. Setiap irisan mewakili bagian dari keseluruhan, dan ukuran irisan tersebut menunjukkan proporsi relatif dari kategori tersebut terhadap total. Pie Chart sangat berguna untuk menggambarkan distribusi data dalam konteks keseluruhan.
Kapan menggunakan Pie Chart:
1. Proporsi atau Persentase: Pie Chart digunakan ketika tujuan utama adalah untuk menunjukkan proporsi atau persentase dari berbagai kategori dalam satu set data.
2. Data Kategorikal: Pie Chart cocok untuk data kategorikal, di mana setiap kategori memiliki arti khusus dan tidak kontinu.
3. Jumlah Kategori Terbatas: Pie Chart lebih efektif ketika jumlah kategori yang ditampilkan tidak terlalu banyak (biasanya kurang dari 6-7 kategori) agar tetap mudah dibaca.
4. Visualisasi yang Sederhana: Pie Chart memberikan visualisasi yang sederhana dan mudah dipahami, sehingga cocok untuk presentasi data kepada audiens yang tidak terlalu teknis.
Contoh penggunaan Pie Chart:
- Menampilkan distribusi penjualan per kategori produk.
- Menunjukkan persentase pangsa pasar dari berbagai perusahaan.
- Menggambarkan proporsi anggaran dalam berbagai departemen.

```python
# Tabel bantu
pm = sales['payment_method'].value_counts().reset_index()
pm
```

```python
# Pie chart dengan gradasi hijau
plt.figure(figsize=(8,8))

pm_sorted = pm.sort_values('count')

# Gradasi hijau dari terang ke gelap
colors = plt.cm.Reds(np.linspace(0.3, 0.9, len(pm_sorted)))

plt.pie(pm_sorted['count'], 
         labels=pm_sorted['payment_method'], 
         autopct='%.1f%%', # persentase
         startangle=90, # mulai dari atas
         colors=colors,
         explode=[0.05]*len(pm_sorted)) # meledakkan irisan
plt.title('Distribusi Metode Pembayaran')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
print("========================================================================")
print(tips.info())
```

```python
print("Rata-rata tip:", tips['tip'].mean())
print("Rata-rata total bill:", tips['total_bill'].mean())
print("Tip tertinggi:", tips['tip'].max())
print("Bill termahal:", tips['total_bill'].max())
```

```python
tips['day'].value_counts()
```

```python
plt.figure(figsize=(8,5))
bar_plot = sns.barplot(data=tips, x='day', y='tip', palette='viridis', errorbar=None)

# Tambahkan label nilai di atas setiap bar
for container in bar_plot.containers:
    bar_plot.bar_label(container, 
                       fmt='%.2f', 
                       padding=1, 
                       fontsize=10, 
                       fontweight='bold')

plt.title('Rata-rata Tips per Hari')
plt.ylabel('Tips ($)')
plt.show()
```

```python
# Total bill berdasarkan lunch vs dinner
plt.figure(figsize=(6,5))
sns.barplot(data=tips, 
            x='time', 
            y='total_bill', 
            palette='coolwarm', 
            errorbar=None)
plt.title('Rata-rata Total Bill: Lunch vs Dinner')
plt.ylabel('Total Bill ($)')
plt.show()
```

## Insight dinner

```python
# Waktu makan yang paling orang sukai
plt.figure(figsize=(6,6))
time_counts = tips['time'].value_counts()

plt.pie(time_counts.values, 
         labels=time_counts.index, 
         autopct='%1.1f%%',
         colors=['lightblue', 'lightcoral'])
plt.title('Kapan Orang Lebih Suka Makan?')
plt.show()
```

```python
# Cek data dinner
dinner_data = tips[tips['time'] == 'Dinner']
lunch_data = tips[tips['time'] == 'Lunch']

print(f"Dinner customers: {len(dinner_data)} orang (76.6%)")
print(f"Rata-rata party size dinner: {dinner_data['size'].mean():.1f} orang")
print(f"Rata-rata total bill dinner: ${dinner_data['total_bill'].mean():.2f}")
print()
print(f"Lunch customers: {len(lunch_data)} orang (23.4%)")
print(f"Rata-rata party size lunch: {lunch_data['size'].mean():.1f} orang")
print(f"Rata-rata total bill lunch: ${lunch_data['total_bill'].mean():.2f}")
```

```python
# Filter data dinner
dinner_data = tips[tips['time'] == 'Dinner']

# Cek distribusi gender di dinner
gender_dinner = dinner_data['sex'].value_counts()
print("Gender distribution saat dinner:")
print(gender_dinner)
print()

# Rata-rata party size berdasarkan gender saat dinner
avg_size_gender = dinner_data.groupby('sex')['size'].mean()
print("Rata-rata party size berdasarkan gender (dinner):")
print(f"Male: {avg_size_gender['Male']:.1f} orang")
print(f"Female: {avg_size_gender['Female']:.1f} orang")
print()

# Cek party size 2 orang (couple) saat dinner
couple_dinner = dinner_data[dinner_data['size'] == 2]
print(f"Dinner dengan party size 2 (couple): {len(couple_dinner)} dari {len(dinner_data)} dinner")
print(f"Persentase couple dinner: {len(couple_dinner)/len(dinner_data)*100:.1f}%")
```

```python
plt.figure(figsize=(8,5))
dinner_size_counts = dinner_data['size'].value_counts().sort_index()

bars = plt.bar(dinner_size_counts.index, dinner_size_counts.values, 
               color='lightcoral', alpha=0.8)

# Tambahkan label
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

plt.title('Distribusi Party Size saat Dinner')
plt.xlabel('Party Size (jumlah orang)')
plt.ylabel('Jumlah Dinner')
plt.xticks(dinner_size_counts.index)
plt.show()
```

```python
# Crosstab gender vs party size untuk dinner
print("CROSSTAB: Gender vs Party Size (Dinner only)")
print("="*45)
gender_size_dinner = pd.crosstab(dinner_data['sex'], dinner_data['size'])
print(gender_size_dinner)
print()

# Analisis couple (size 2) berdasarkan gender
couple_gender = couple_dinner['sex'].value_counts()
print("Gender yang bayar saat couple dinner (size 2):")
print(couple_gender)
print(f"Male bayar: {couple_gender['Male']} kali ({couple_gender['Male']/len(couple_dinner)*100:.1f}%)")
print(f"Female bayar: {couple_gender['Female']} kali ({couple_gender['Female']/len(couple_dinner)*100:.1f}%)")
```

Insight:
- DESSERT SHARING untuk couple
- ROMANTIC ATMOSPHERE (dim light, candle)
- COUPLE PACKAGE (2 main course + 1 dessert)

## Smoker vs Non-smoker

```python
# Cek spending pattern
spending = tips.groupby('smoker').agg({
    'total_bill': ['mean', 'sum', 'count'],
    'tip': 'mean'
}).round(2).reset_index()

display(spending)
```

```python
# Potensial pasar
non_smoker_count = len(tips[tips['smoker'] == 'No'])
smoker_count = len(tips[tips['smoker'] == 'Yes'])

print(f"Target Market:")
print(f"Non-perokok: {non_smoker_count} orang ({non_smoker_count/len(tips)*100:.1f}%)")
print(f"Perokok: {smoker_count} orang ({smoker_count/len(tips)*100:.1f}%)")
```

```python
plt.figure(figsize=(8,5))
avg_spending = tips.groupby('smoker')['total_bill'].mean()

bars = plt.bar(['Perokok', 'Non-Perokok'], avg_spending.values, 
               color=['salmon', 'lightgreen'], alpha=0.8)

for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'${height:.2f}', ha='center', va='bottom', fontweight='bold')

plt.title('Rata-rata Spending per Kunjungan')
plt.ylabel('Total Bill ($)')
plt.ylim(0, max(avg_spending.values) * 1.2)
plt.show()
```

```python
# Pie chart: Market share potential
plt.figure(figsize=(10,5))

# Subplot 1: Jumlah customer
plt.subplot(1,2,1)
customer_counts = tips['smoker'].value_counts()
plt.pie(customer_counts.values,
         labels=['Non-Perokok', 'Perokok'],
         autopct='%1.1f%%',
         colors=['lightgreen', 'salmon'],
         explode=[0.1, 0])
plt.title('Total Customer antara Perokok vs Non-Perokok')


plt.subplot(1,2,2)
revenue_potential = tips.groupby('smoker')['total_bill'].sum()
plt.pie(revenue_potential.values,
         labels=['Non-Perokok', 'Perokok'],
         autopct='%1.1f%%',
         colors=['lightgreen', 'salmon'],
         explode=[0.1, 0])
plt.title('Total spend antara Perokok vs Non-Perokok')

plt.tight_layout()
plt.show()
```

```python
non_smoker_avg = avg_spending['No']
smoker_avg = avg_spending['Yes']
gap = smoker_avg - non_smoker_avg

print(f"Spending Gap: ${gap:.2f} per kunjungan")
print(f"Non-perokok spend {gap/smoker_avg*100:.1f}% lebih sedikit")
print(f"Target Market: {non_smoker_count} orang (61.9% dari total)")
print()
```

Strategi bisnis:
- Jual cemilan sehat seharga 2-3$
- Fokus pada pemasaran untuk non-perokok
- Cemilan tersebut bisa menaikan spending mereka
- Market besar 151 vs 93
