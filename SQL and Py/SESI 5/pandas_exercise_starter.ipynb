## Create Dataframe üî≥

### Question 1

Membuat dataframe dengan 3 kolom yaitu x,y dan z. memiliki data berurutan dari setiap kolom angka 1 sampai 15

**Answer** ‚¨áÔ∏è 

```python
df = pd.DataFrame({
    'x': range(1, 6),
    'y': range(6, 11),
    'z': range(11, 16)
})

df
```

### Question 2
Menambahkan kolom 'v' yang memiliki nilai data kebawah yaitu 16-20

**Answer** ‚¨áÔ∏è 

```python
df['v'] = np.arange(16, 21)
df
```

## Selection & indexing ü§ö

### Question 1
Dapatkan kolom dari 'x' dan 'y'

**Answer** ‚¨áÔ∏è 

```python
df[['x', 'y']]
```

### Question 2
Dapatkan data dengan index 2 dan hanya kolom 'x' dan 'y' saja

**Answer** ‚¨áÔ∏è 

```python
df.loc[[2], ['x', 'y']]
```

### Question 3
Seleksi data yang memiliki nilai lebih dari 13

**Answer** ‚¨áÔ∏è 

```python
df[df > 13]
```

## Data Manipulation üßô

```python
df
```

### Question 1
Buat kolom baru nama 'f' yaitu dengan menggunakan nilai dari kolom 'v' apakah lebih besar dari 18 atau tidak

**Answer** ‚¨áÔ∏è 

```python
def fungsi(f):
    if f > 18:
        return 'High'
    else:
        return 'Low'

df['f'] = df['v'].apply(fungsi)
df
```

### Question 2
Hapus kolom 'y'

**Answer** ‚¨áÔ∏è 

```python
df.drop(columns=['y'], inplace=True)
df
```

### Question 3
Urutkan data berdasarkan kolom 'v' secara descending

**Answer** ‚¨áÔ∏è 

```python
df.sort_values('v', ascending=False)
```
