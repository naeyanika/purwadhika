## Menambahkan kolom dan baris pada DataFrame

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dataMember = {'Nama': ['Jaemin', 'Jeno', 'Haechan', 'Mark', 'Chenle', 'Renjun', 'Jisung'],
        'Umur': [26, 25, 24, 27, 21, 23, 22],
        'Posisi': ['Rapper', 'Rapper', 'Vocalist', 'Rapper', 'Vocalist', 'Vocalist', 'Rapper'],
        'Alamat': ['Seoul', 'Seoul', 'Seoul', 'Seoul', 'Seoul', 'Seoul', 'Seoul']}
df3 = pd.DataFrame(dataMember)
df3
```

```python
# menambahkan kolom di bagian akhir dengan syarat jumlah data sesuai dengan jumlah baris
df3['Tinggi'] = [175, 178, 176, 180, 182, 177, 174]
df3['Alamat'] = ['Seoul', 'Seoul', 'Incheon', 'Busan', 'Seoul', 'Seoul', 'Seoul'] # mengubah/replace isi kolom Alamat
df3
```

```python
# Menambahkan kolom di bagian yang diinginkan
df3.insert(2, 'Berat', [60, 65, 55, 70, 50, 58, 62]) # (index kolom, nama kolom, data kolom)
df3
```

```python
# Menambahkan baris
df3.loc[7] = ['Jungwoo', 24, 68, 'Vocalist', 'Seoul', 180] # (index baris, data baris)
df3
```

```python
# Menambahkan kolom dari 2 kolom yang sudah ada
df3['BMR'] = 66 + (13.7 * df3['Berat']) + (5 * df3['Tinggi']) - (6.8 * df3['Umur'])
df3
```

## Menghapus Kolom
**Notes:**
1. Axis = 0 : Merujuk ke baris
2. Axis = 1 : Merujuk ke kolom
3. inplace = True : Mengubah DataFrame asli (permanen)

```python
df3
```

```python
## Menghapus kolom Tinggi_Berat tidak permanen
df3.drop('BMR', axis=1, inplace=False) # tidak permanen
df3
```

```python
df3.drop(columns=['Tinggi_Berat'], inplace=False) # tidak permanen
```

## Set dan Reset Index
**Notes:**
1. Set Index : Mengubah kolom menjadi index
2. Reset Index : Mengubah index menjadi kolom biasa

```python
# Set nama menjadi index
df3.set_index('Nama') # tidak permanen baik pakai inplace=false atau tidak
```

```python
df3.set_index('Nama', inplace=True) # permanen
df3
```

```python
df3.reset_index(inplace=True) # permanen
df3
```

## Sorting
**Notes:**
1. sort_values() : Mengurutkan berdasarkan nilai pada kolom tertentu
2. sort_index() : Mengurutkan berdasarkan index

```python
## Sort berdasarkan value kolom Umur secara descending
df3.sort_values('Umur', ascending=False, inplace=True) # permanen
df3
```

```python
## Sort berdasarkan index secara ascending
df3.sort_index(ascending=True, inplace=True) # permanen
df3
```

## Apply Fungsi ke DataFrame
**Notes:**
1. apply() : Menerapkan fungsi ke DataFrame
2. lambda : Fungsi anonim (fungsi tanpa nama)

```python
## Membuat fungsi sederhana
def pangkat(x):
    data = x ** 3
    return data

## Apply Fungsi ke DataFrame
df3['Umur_Pangkat_3'] = df3['Umur'].apply(pangkat) # Nilai yang muncul adalah hasil dari fungsi pangkat
df3.drop(columns=['Umur_Pangkat_2'], inplace=True) # menghapus kolom Umur_Pangkat_2
df3
```

```python
# Fungsi klasifikasi tinggi badan
def tinggi(tb):
    if tb >= 180:
        return 'Tinggi'
    elif tb >= 175:
        return 'Sedang'
    else:
        return 'Pendek'

df3['Klasifikasi Tinggi'] = df3['Tinggi'].apply(tinggi)
df3
```

### Missing Values
---
1. isnull() : Menampilkan nilai null pada DataFrame
2. notnull() : Menampilkan nilai tidak null pada DataFrame
3. dropna() : Menghapus baris/kolom dengan nilai null
4. fillna() : Mengisi nilai null dengan nilai tertentu

```python
# menampilkan nilai null pada DataFrame
titanic.isna().sum() 
```

```python
# Cek missing value berdasarkan persen
titanic.isna().sum() / len(titanic) * 100
```
