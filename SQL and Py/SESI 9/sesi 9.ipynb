# Statistik

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = np.random.normal(loc=50, scale=10, size=1000)
```

```python
#KDE Plot
plt.figure(figsize=(10,10))
sns.kdeplot(data, color='skyblue')
plt.title('Distribusi Data Normal')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Kolmogorov-Smirnov Test
from scipy.stats import kstest
kstest(data, 'norm') # ketika lebih kecil dari 0.05 maka data tidak berdistribusi normal (pvalue < 0.05 atau 5%)
```

```python
# Shapiro test
from scipy.stats import shapiro

shapiro(data) # ketika lebih kecil dari 0.05 maka data tidak berdistribusi normal (pvalue < 0.05 atau 5%)
```

```python
# D agustino test
from scipy.stats import normaltest
normaltest(data) # ketika lebih kecil dari 0.05 maka data tidak berdistribusi normal (pvalue < 0.05 atau 5%)
```

## Pakem uji normalitas

- ketika value < 0.05 maka data tidak berdistribusi normal (pvalue < 0.05 atau 5%)
- ketika value > 0.05 maka data berdistribusi normal (pvalue > 0.05 atau 5%)
- pakai mean jika data berdistribusi normal
- pakai median jika data tidak berdistribusi normal
